```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
```

```python
imagem  = cv2.imread('MM.jpg')
cv2_imshow(imagem)
```

```python
print(imagem.shape[0])
print(imagem.shape[1])
imagem
```

```python
imagem_cinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)
cv2_imshow(imagem_cinza)
```

```python
cv2.imwrite('imagem_cinza.jpg', imagem_cinza)
```

```python
novo_tamanho = (500, 500)
imagem_resolucao = cv2.resize(imagem_cinza, novo_tamanho, interpolation = cv2.INTER_AREA)
cv2_imshow(imagem_resolucao)
```

```python

print(imagem_resolucao.shape[0])
print(imagem_resolucao.shape[1])
cv2.imwrite('imagemCinza.jpg', imagem_resolucao)
```

```python
qtd = 32
img_niveisCinza = np.uint8(imagem_cinza/qtd)*qtd
cv2_imshow(img_niveisCinza)
```

```python

cv2.imwrite('img_niveisCinza.jpg', img_niveisCinza)
```
